# ⭐ distibution plot
-----

distplot() is a function that shows how data is spread out.
It combines a histogram (bar graph) and a KDE curve (smooth line) to display the distribution of a single variable.

✅ Use:
To see the shape of the distribution of a dataset (e.g., normal, skewed).

```python
# distibutions plot
vis1=sns.distplot(store['InternetUsers'])  # univariate analysis - plot graph using 1 variable
plt.show()
```

```python
vis1=sns.distplot(store['InternetUsers'],bins=10)  # univariate analysis - plot graph using 1 variable
plt.show()
```

# ⭐ Boxplot
----------

🔹 Definition:

boxplot() is used to display the distribution of data based on five summary statistics: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. It also highlights outliers.

✅ Use:
To visualize the spread and skewness of data and to detect outliers.

```python
# Box plot
vis2 = sns.boxplot (data=store,x="IncomeGroup",y='BirthRate') # Bi- variate analysis
plt.show()
```

```python
# dot in graph are outliers
# line inside the box is average
# box line - 25% quartile - niche se box line
# box line - 75% quartile - upar vala box line
```

# ⭐ lmplot
-----
sns.lmplot():
A figure-level function in Seaborn that combines scatter plots with linear regression fits, optionally conditioned on additional categorical variables for faceting and coloring.

```python
# linear model plot
vis3=sns.lmplot(data=store,x='InternetUsers',y='BirthRate')  
plt.show()
```

```python
# linear model plot
vis3=sns.lmplot(data=store,x='InternetUsers',y='BirthRate',fit_reg=False)
plt.show()
```

```python
# linear model plot
vis3=sns.lmplot(data=store,x='InternetUsers',y='BirthRate',fit_reg=False,hue = 'IncomeGroup')
plt.show()
```
